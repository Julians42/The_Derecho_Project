```python
### Import basic libraries
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
#%matplotlib inline
import time
import glob
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
# We define the following metrics for growing and killing winds 
def kill_winds(windspeed, threshold):
    if windspeed <= threshold:
        return 0
    else:
        return windspeed - threshold
kill_vectorized = np.vectorize(kill_winds)

def grow_winds(windspeed, threshold):
    if windspeed <= 2:
        return windspeed
    elif (windspeed>2) & (windspeed<threshold-2):
        return 2
    elif (windspeed <= threshold) & (windspeed > threshold-2):
        return threshold - windspeed
    else:
        return 0
grow_vectorized = np.vectorize(grow_winds)

#GDD and KDD functions
def KDD_fn(x, threshold):
    if x < threshold:
        return 0
    if x >= threshold:
        return x - threshold
KDD_vectorized = np.vectorize(KDD_fn)
def GDD_fn(x, t1, t2):
    if x <t1:
        return 0
    if t1<=x<t2:
        return x-t1
    if x>=t2:
        return 24
GDD_vectorized = np.vectorize(GDD_fn)
```

```python
x_s = np.linspace(0,25, 2000)
t_s = [10, 15, 20]
color = ['orange', 'green', 'blue']
for i in range(len(t_s)):
    y_grow = [grow_winds(x, t_s[i]) for x in x_s]
    y_kill = [kill_winds(x, t_s[i]) for x in x_s]
    label = "Threshold = {} m/s".format(t_s[i])
    plt.plot(x_s, y_grow, np.negative(y_kill)]), linewidth=3, label=label, color = color[i], alpha = 0.5)
#plt.axhline(y=0, color='r', linewidth=2)
#plt.plot(months, np.zeros(len(months)), color = 'grey', linewidth = 2.5, alpha = 0.7, linestyle = 'dotted')
plt.axhline(y = 0, color = 'grey', linewidth = 2.5, alpha = 0.7, linestyle = 'dotted')
plt.title("Estimated Wind Response Function")
plt.xlabel("Windspeed (m/s)")
plt.ylabel("Response Weight")
plt.legend(loc = 'best');
```

```python
def compute_KD(filenames = filenames, height = 'wind_10ms', threshold = 15):
    """ Takes a list of wind data files and returns 
         - location based metrics: year, state, county
         - log-yield
         - sum of windspeed, windspeed^2, windspeed^3
         - GDD, KDD, KW, and GW derived from the data
        Returns in a big dataframe
    """
    data_plt = []
    #hourly_winds = [] # array of arrays of windspeeds 
    for i in range(len(filenames)):
    #for i in range(1):
        wind_i = pd.read_csv(filenames[i], low_memory = False)
        county_name = filenames[i].split("_")[-2].upper() # extract county name
        state_name = filenames[i].split("_")[-3].upper() # extract state name
        county_data = county_yield.loc[(county_yield['County'] == county_name) \
                                       & (county_yield['State']==state_name)] # kansas yeild data
        c_year = county_data["Year"].unique() 
        w_year = wind_i["Year"].unique()
        years = np.intersect1d(c_year, w_year) # finds all common values -essentially just the wind years 
        # filter years 
        county_gs = county_data[county_data["Year"].isin(years)] 
        wind_gs = wind_i[wind_i["Year"].isin(years)]
        wind_gs = wind_gs[(wind_gs["Month"] >= 4) & (wind_gs["Month"] <= 10)] # growing season
        for y in years:
            wind_raw = wind_gs.loc[wind_gs["Year"] == y][height].values # grab wind data
            county_gs_y = county_gs.loc[county_gs["Year"] == y] # get yield data
            #w_gs_y = wind_gs_y[height] # compute for specific height
            average_winds = np.mean(wind_raw)
            average_winds2 = np.mean(wind_raw**2) # sum of squares
            average_winds3 = np.mean(wind_raw**3) # sum of cubes
            # Compute the killing and growing wind values over wind_raw
            kill_w = np.sum(kill_vectorized(wind_raw, threshold))#np.sum([kill_winds(t,threshold) for t in wind_raw])
            grow_w = np.sum(grow_vectorized(wind_raw, threshold)) #np.sum([grow_winds(t,threshold) for t in wind_raw])
            # compute KDD and GDD
            temperature_raw = wind_gs.loc[wind_gs["Year"] == y]["temp_10"].values
            KDD = np.sum(KDD_vectorized(temperature_raw, 30)) #np.sum([KDD_fn(t, 30) for t in temperature_raw])
            GDD = np.sum(GDD_vectorized(temperature_raw, 8, 30)) #np.sum([GDD_fn(t, 8, 30) for t in temperature_raw])
            ##########NEW#########
            wind_values = [GDD, KDD, average_winds, average_winds2, average_winds3, kill_w, grow_w]
            wind_gs_y = wind_gs.loc[wind_gs["Year"] == y]
            months = wind_gs_y["Month"].unique()
            #months = wind_raw["Month"].unique()
            # Looping through all the months in the growing season to compute the damage per month
            for m in months:
                wind_gs_m = wind_gs_y.loc[wind_gs_y["Month"] == m] #wind_raw.loc[wind_raw["Month"] == m]
                w_gs_m = wind_gs_m[height].values
                killing_winds_m = np.sum(kill_vectorized(w_gs_m, threshold))
                growing_winds_m = np.sum(grow_vectorized(w_gs_m, threshold))
                wind_values.append(killing_winds_m) 
                wind_values.append(growing_winds_m)
            #########NEW###########
            # year, state county, ln_yield, GDD, KDD  - merge with average wind speeds
            data = np.append(county_gs_y.iloc[:,[1,3,5,12,16,20]].values, wind_values).tolist() #[GDD, KDD, average_winds, average_winds2, average_winds3, kill_w, grow_w]).tolist()
            data_plt.append(data) # append to data
    data_plt = np.array(data_plt)
    df_plt = pd.DataFrame(data_plt, columns = ['Year', 'State', 'County', 'lnyield', 'PPT','PPT2','GDD', 'KDD', 'MeanWindspeed', "Windspeed2", "Windspeed3", "KW","GW",'KW4','GW4','KW5','GW5','KW6','GW6','KW7','GW7','KW8','GW8','KW9','GW9','KW10','GW10'])
    # convert to correct datatype
    covars = ['lnyield','GDD','KDD','PPT','PPT2','KW','GW','MeanWindspeed','Windspeed2','Windspeed3','KW4','GW4','KW5','GW5','KW6','GW6','KW7','GW7','KW8','GW8','KW9','GW9','KW10','GW10']
    for d in covars:
        df_plt[d] = df_plt[d].astype('float64')
    #df_plt['lnyield'] = df_plt['lnyield'].astype('float64')
    #df_plt['PPT'] = df_plt['PPT'].astype('float64')
    #df_plt['PPT2'] = df_plt['PPT2'].astype('float64')
    #df_plt['GDD'] = df_plt['GDD'].astype('float64')
    #df_plt['KDD'] = df_plt['KDD'].astype('float64')
    #df_plt['KW'] = df_plt['KW'].astype('float64')
    #df_plt['GW'] = df_plt['GW'].astype('float64')
    #df_plt['MeanWindspeed'] = df_plt['MeanWindspeed'].astype('float64')
    #df_plt['Windspeed2'] = df_plt['Windspeed2'].astype('float64')
    #df_plt['Windspeed3'] = df_plt['Windspeed3'].astype('float64')
    #df_plt[['lnyield','PPT','PPT2','GDD','KDD','KW','GW','MeanWindspeed','Windspeed2','Windspeed3']]= \
    #                            df_plt.drop(['Year','State','County'], axis=1).astype('float64')
    df_plt['Year'] = df_plt['Year'].astype('int32')
    
    df_mean = df_plt.groupby(by=["Year","State","County"], axis=0, as_index=False).mean()
    df_mean['ST_CO'] = df_mean['State']+'_'+df_mean['County'] # get unique county identifiers
    df_mean['PPT_SQUARE'] = df_mean['PPT']**2 # add precipitation squared variable for regression
    return df_mean
```

```python
def extract_params(model, threshold):
    dct = {"threshold":threshold,"Coef KW4":model.params["KW4"], "KW4 Pvalue":model.pvalues["KW4"],"KW4 SE":model.bse['KW4'], \
           "Coef KW5":model.params["KW5"], "KW5 Pvalue":model.pvalues["KW5"],"KW5 SE":model.bse['KW5'], \
           "Coef KW6":model.params["KW6"], "KW6 Pvalue":model.pvalues["KW6"],"KW6 SE":model.bse['KW6'], \
           "Coef KW7":model.params["KW7"], "KW7 Pvalue":model.pvalues["KW7"],"KW7 SE":model.bse['KW7'], \
           "Coef KW8":model.params["KW8"], "KW8 Pvalue":model.pvalues["KW8"],"KW8 SE":model.bse['KW8'], \
           "Coef KW9":model.params["KW9"], "KW9 Pvalue":model.pvalues["KW9"],"KW9 SE":model.bse['KW9'], \
           "Coef KW10":model.params["KW10"],"KW10 Pvalue":model.pvalues["KW10"],"KW10 SE":model.bse['KW10'], \
           "rsquared":model.rsquared, 'adj_rsquared':model.rsquared_adj, \
           "Coef PPT":model.params["PPT"], "PPT Pvalue":model.pvalues["PPT"], \
           "PPT SE":model.bse['PPT'], "Coef PPT_SQUARE":model.params["PPT2"], \
           "PPT_SQUARE Pvalue":model.pvalues["PPT2"], "PPT_SQUARE SE":model.bse['PPT2']}
    return dct
#params_15 = extract_params(model, 15)
#print(params_15)
#print(params_15['adj_rsquared'])
```

```python
start_time = time.time()
thresholds = np.arange(0, 26, 1)#[2, 15]#np.arange(15, 17, 1)
params_df = pd.DataFrame([], columns = ['threshold', 'Coef KW4', 'KW4 Pvalue', 'KW4 SE', 'Coef KW5','KW5 Pvalue', 'KW5 SE', 'Coef KW6', 'KW6 Pvalue', 'KW6 SE', 'Coef KW7','KW7 Pvalue', 'KW7 SE', 'Coef KW8', 'KW8 Pvalue', 'KW8 SE', 'Coef KW9','KW9 Pvalue', 'KW9 SE', 'Coef KW10', 'KW10 Pvalue', 'KW10 SE','rsquared', 'adj_rsquared', 'Coef PPT', 'PPT Pvalue', 'PPT SE','Coef PPT_SQUARE', 'PPT_SQUARE Pvalue', 'PPT_SQUARE SE'])
for t in thresholds:
    print(t)
    df_wind_i = compute_KD(threshold = t)
    df_wind_i['PPT_SQUARE'] = df_wind_i['PPT']**2
    data_df_resid = pd.DataFrame([], columns = ['Year', 'State', 'County', 'lnyield', 'PPT','PPT2','GDD', 'KDD', 'MeanWindspeed', "Windspeed2", "Windspeed3", "KW","GW",'KW4','GW4','KW5','GW5','KW6','GW6','KW7','GW7','KW8','GW8','KW9','GW9','KW10','GW10'])
    detrends = ['lnyield', 'PPT','PPT2','GDD', 'KDD', 'MeanWindspeed', "Windspeed2", "Windspeed3", "KW","GW",'KW4','GW4','KW5','GW5','KW6','GW6','KW7','GW7','KW8','GW8','KW9','GW9','KW10','GW10']
    states = list(df_wind_i['State'].unique())
    for s in states:
        data_s = df_wind_i.loc[df_wind_i['State'] == s].copy()
        counties = list(data_s['County'].unique())
        for i in range(len(counties)):
            data_i = data_s.loc[data_s['County'] == counties[i]].copy()
            covariates = data_i['Year'].astype(float).values.reshape(-1, 1)
            model = LinearRegression()
            for d in detrends:
                ys = data_i[d].astype(float)
                model.fit(covariates, ys)
                y_preds = model.predict(covariates)
                residuals = ys - y_preds
                data_i.loc[:, d] = residuals
            data_df_resid = data_df_resid.append(data_i)
    model_i = ols('lnyield ~ C(Year) + C(County) + GDD + KDD + KW4+GW4+KW5+GW5+KW6+GW6+KW7+GW7+KW8+GW8+KW9+GW9+KW10+GW10 + PPT + PPT2', data=data_df_resid).fit() #
    #print(model_i.summary())
    #model_i = ols('lnyield ~ C(Year) + C(County) + GDD + KDD + KW + GW + PPT + PPT_SQUARE', data=df_wind_i).fit()
    params_i = extract_params(model_i, t)
    params_df_i = pd.DataFrame(params_i, index = [0])
    #print(params_df_i)
    params_df = params_df.append(params_df_i)
print(time.time() - start_time)
```

```python
fig, ax = plt.subplots(1,2,figsize = (12,4))
fig.suptitle('OLS with County and Year Fixed Effects with 95% C.I.')
ax[0].scatter(data_df_resid['KW'], data_df_resid['lnyield'], color = 'grey', alpha = 0.1)
def kw_regress(x):
    return model.params['KW']*x
def kw_lower(x):
    return (model.params['KW'] - 2*model.bse['KW'])*x
def kw_upper(x):
    return (model.params['KW'] + 2*model.bse['KW'])*x
xlim = np.arange(np.min(data_df_resid['KW']), np.max(data_df_resid['KW']), 0.1)
lower_bound = kw_lower(xlim)
upper_bound = kw_upper(xlim)
label = "$r^2 = $" + str(round(model.rsquared_adj,5))
ax[0].plot(xlim, kw_regress(xlim), color = 'green', linewidth = 3.5, label = label)
ax[0].fill_between(xlim, lower_bound, upper_bound, color = 'green', alpha = 0.2)
ax[0].set_title('Linear Regression')
ax[0].set_xlabel('Detrended Killing Wind (m/s)')
ax[0].set_ylabel('Detrended Log Yield')
ax[0].legend(loc = 'best')

#polynomial plot
def kw_poly(x):
    return model_poly.params['MeanWindspeed']*x + model_poly.params['Windspeed2']*x**2 + model_poly.params['Windspeed3']*x**3
def kw_poly_lower(x):
    c1 = model_poly.params['MeanWindspeed'] - 2*model_poly.bse['MeanWindspeed']
    c2 = model_poly.params['Windspeed2'] - 2*model_poly.bse['Windspeed2']
    c3 = model_poly.params['Windspeed3'] - 2*model_poly.bse['Windspeed3']
    return c1*x + c2*x**2 + c3*x**3
def kw_poly_upper(x):
    c1 = model_poly.params['MeanWindspeed'] + 2*model_poly.bse['MeanWindspeed']
    c2 = model_poly.params['Windspeed2'] + 2*model_poly.bse['Windspeed2']
    c3 = model_poly.params['Windspeed3'] + 2*model_poly.bse['Windspeed3']
    return c1*x + c2*x**2 + c3*x**3
ax[1].scatter(data_df_resid['MeanWindspeed'], data_df_resid['lnyield'], color = 'grey', alpha = 0.1, label = 'Detrended')
xlim2 = np.arange(np.min(data_df_resid['MeanWindspeed']), np.max(data_df_resid['MeanWindspeed']), 0.1)
lower_poly = kw_poly_lower(xlim2)
upper_poly = kw_poly_upper(xlim2)
label = "$r^2 = $" + str(round(model_poly.rsquared_adj,5))
ax[1].plot(xlim2, kw_poly(xlim2), color = 'orange', linewidth = 3.5, label = label)
ax[1].fill_between(xlim2, lower_poly, upper_poly, color = 'orange', alpha = 0.2)
ax[1].set_title('Cubic Polynomial Regression')
ax[1].set_xlabel('Detrended Mean Wind Speed during Growing Season (m/s)')
ax[1].set_ylabel('Detrended Log Yield');
ax[1].legend(loc = 'best');
```
