```python
def compress_file(filename):
    wind = pd.read_csv(filename, low_memory=False)
    wind2 = wind.groupby(["Year","Month","Day","Hour"], as_index=False).max()[["Year","Month","Day","Hour", "wind speed at 10m (m/s)"]]
    wind2['air temperature at 10m (C)'] = wind.groupby(["Year","Month","Day","Hour"], as_index=False).mean()[['air temperature at 10m (C)']]
```
